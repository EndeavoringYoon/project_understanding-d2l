Our goal is to discover patterns that capture regularities in the underlying population from which our training set was drawn.

This problem-how to discover patterns that generalize-is the fundamental problem of machine learning

The danger is that when we train models, we access just a small sample of data.

The phenomenon of fitting our training data more closely than we fit the underlying distribution is called overfitting, and the techniques used to combat overfitting are called regularization.

# 4.4.1 Training Error and Generalization Error

Training Error || Generalization Error

* The training error is the error of our model as calculated on the training dataset.

* generalization error is the expectation of our model’s error were we to apply it to an infinite stream of additional data examples drawn from the same underlying data distribution as our original sample.

  we can never calculate the generalization error exactly. That is because the stream of infinite data is an imaginary object.
  In practice, we must estimate the generalization error by applying our model to an independent test set constituted of a random selection of data examples that were withheld from our training set.
  
generalization error는 무한한 데이터를 가질 수 없기 때문에 정확하게 측정할 수 없음
그래서 training set과 겹치지 않는, random하게 선택된 data를 바탕으로 test set을 구성하여 우리 model을 적용하고, 그 결과를 바탕으로 generalization error를 측정함.

## 4.4.1.1 Statistical Learning Theory

Glivenko and Cantelli Theorem, 혹은 그에서 발전한 Vapinik and Chervonenkis theorem에서 training error가 generalization error에 수렴하는 rate를 얻어냄

standard supervised learning setting에서는 training data와 test data가 동일한 분포를 가지는 독립적인 data라고 가정함.(are drawn **independently** from **identical distributions** a.k.a. **i.i.d. assumption**)

이 i.i.d assumption 자체가 가지는 문제점도 있음
(data간의 독립성을 보장할 수 없거나, test,train data가 가지는 분포가 다를 경우 등. 여기서는 환자 정보를 가져오는 병원의 차이, 트위터 주제 분류 시스템의 시간적 의존성, 얼굴인식 시스템의 train-대학, test-요양원에서 오는 표본 오류 등)



i.i.d. assumption에서 발생하는 문제는 차후의 chapters에서 분석하자. 하지만 i.i.d. assumption 자체를 수용하더라도 generalization의 이해는 다른 문제이다.

## 4.4.1.2 Model Complexity

simple models& abundant data ------(expect)------>generalization error ~ training error

complex models & fewer examples ------(expect)----->training error go down, but generalization gap to grow.

complex matter(ex. model with more parameters, or parameters which can take a wider range of values: more complex )

training iterations가 많을 수록 더 복잡하다고 생각할 수 있지만 fewer training iterations가 더 복잡함

경험적 법칙으로, arbitary facts를 쉽게 설명할 수 있는 model을 복잡하다고 봄. 제한된 표현력으로 많은 데이터를 다룰 수 있을 수록 사실에 가까울 것이라 예상되기 때문.

이 section에서 주목할, model class의 generaliability에 영향을 줄 factor들:

1. number of tunable parameters.
    tunable parameters(=degrees of freedom)의 수가 많을수록 model은 overfitting되기 쉬워짐

2. The values taken by parameters.
    weights가 values의 큰 범위를 차지한다면 model은 overfitting되기 쉬워짐

3. The number of training examples.
    표본 수가 적을수록 overfitting하기 쉬워짐

# 4.4.2 Model Selection

machine learning에서는 몇몇 모델 후보들의 계산을 진행해보고 final model을 선정하는데, 이 과정을 model selection이라고 함.

이는 서로 완전히 다른 모델을 비교할 때도 있지만, 같은 class의 모델들을 다른 hyperparameter settings로 비교하는 경우도 포함됨.

예시로 MLP에서는 hidden layers의 개수나 unit의 개수, activation functions의 종류를 바꿔가며 테스트함. model 중에서 좋은 모델을 결정하기 위해 validation dataset을 사용

## 4.4.2.1 Validation Dataset

정의상 hyperparameter를 결정하기 전까지 test set의 조작을 가할 수는 없음. 그렇다고 test data를 model selection process에 사용하면 test data를 overfit할 위험도 있음.

그러므로 model selection에 test data를 사용할 수 없지만, training data를 model selection에 사용하면 train data의 generalization error를 측정할 수 없음

이 문제에 대한 가장 흔한 대처법은 data를 validation, training, test dataset의 3가지로 분리하는 것임.

이 책의 실험에서는 true test set이 없기 때문에, 각각의 실험에서 나오는 정확도는 true test set accuracy가 아닌 validation accuracy임

## 4.4.2.2 K-Fold Cross-Validation

training data의 부족으로 validation set를 만들 수 없을 경우, K-fold cross-validation을 채택하는 대중적인 해결법이 있음.

원래의 training data를 K개의 non-overlapping subsets로 나누고, model training과 validation은 K-1개의 training subset과 1개의 validation subser을 가지고 진행하는 K번의 실험으로 진행됨.

training과 validation error는 K번의 실험의 값의 평균치가 됨

# 4.4.3 Underfitting or Overfitting?

training, validation errors를 비교할 때 염두에 둘 2가지 상황
1. training error와 validation error가 상관있기를 바라지만 그 둘에는 약간의 차이가 있음.
    만약 model이 training error를 줄일 수 없다면 우리 model은 우리가 modeling하려고 하는 것의 패턴을 파악하기에는 너무 단순하다는 의미일 수 있음
    trainig과 validation error의 generalization gap은 작기 때문에, 더 복잡한 model을 잘 해낼 수 있을 것이라는 생각을 가짐. 이 현상을 underfitting이라고 함.

2. training error가 validation error보다 확연히 작은 경우: overfitting
    overfitting이 항상 나쁜 것은 아님을 염두에 둘 것(딥러닝에서 가장 좋은 예상 model들은 실제보다 training에서 더 나은 성적을 보이기 때문)
    그래서 training과 validation error의 간격보다 validation error에 더 신경을 많이 씀.
    
우리가 overfit이나 underfit을 하는 것은 우리 model의 복잡성이나 이용 가능한 training dataset의 크기에 의존할 수 있음

## 4.4.3.1 Model Complexity

단일 feature x와 실제 label y로 구성된 training data를 가정했을 때, 

$$\hat{y}= \sum_{i=0}^d x^i w_i$$

label y를 추정하기 위해 $d$에 대한 다항식을 찾고자 함.

=linear regression problem(features: x, weights: $w_i$, bias: $w_0$(since $x^0$=1))

linear regression problem이기 때문에, loss function으로 squared error 사용 가능

고차다항식이 더 많은 parameter를 가지고, model function의 선택범위 또한 넓어지기 때문에 고차다항식 함수는 저차다항식 함수보다 더 복잡함.

training dataset을 수정하면 고차다항식 함수는 저차다항식에 비해 항상 더 낮은(최악의 경우 같은) training error를 가짐.
게다가 data example이 분명한 x값을 가질 때, 차수가 data example의 개수와 같은 값을 가지는 다항식은 training set을 완벽하게 만족할 수 있음.

## 4.4.3.2 Dataset Size

Training dataset의 크기가 클수록 generalization error가 감소함.

딥러닝은 데이터 양이 충분할 때 선형 모델보다 좋은 결과를 냄을 항상 염두에 두자.

# 4.4.4 Polynomial Regression

```python
import math
import numpy as np
import torch
from torch import nn
from d2l import torch as d2l
```

## 4.4.4.1 Generating the Dataset

(**$$y = 5 + 1.2x - 3.4\frac{x^2}{2!} + 5.6 \frac{x^3}{3!} + \epsilon \text{ where }
\epsilon \sim \mathcal{N}(0, 0.1^2).$$**)

$\epsilon$과 다항식의 $\frac{x^i}{i!}$ 등은 최적화를 위해 i가 커졌을 때 나올 수 있는 큰 값들을 피하기 위함임.

```python
max_degree = 20  # Maximum degree of the polynomial
n_train, n_test = 100, 100  # Training and test dataset sizes
true_w = np.zeros(max_degree)  # Allocate lots of empty space
true_w[0:4] = np.array([5, 1.2, -3.4, 5.6])

features = np.random.normal(size=(n_train + n_test, 1))
np.random.shuffle(features)
poly_features = np.power(features, np.arange(max_degree).reshape(1, -1))
for i in range(max_degree):
    poly_features[:, i] /= math.gamma(i + 1)  # `gamma(n)` = (n-1)!
# Shape of `labels`: (`n_train` + `n_test`,)
labels = np.dot(poly_features, true_w)
labels += np.random.normal(scale=0.1, size=labels.shape)
```

## 4.4.4.2 Training and Testing the Model

```python
def evaluate_loss(net, data_iter, loss):  #@save
    """Evaluate the loss of a model on the given dataset."""
    metric = d2l.Accumulator(2)  # Sum of losses, no. of examples
    for X, y in data_iter:
        out = net(X)
        y = y.reshape(out.shape)
        l = loss(out, y)
        metric.add(l.sum(), l.numel())
    return metric[0] / metric[1]
```

```python
def train(train_features, test_features, train_labels, test_labels,
          num_epochs=400):
    loss = nn.MSELoss(reduction='none')
    input_shape = train_features.shape[-1]
    # Switch off the bias since we already catered for it in the polynomial
    # features
    net = nn.Sequential(nn.Linear(input_shape, 1, bias=False))
    batch_size = min(10, train_labels.shape[0])
    train_iter = d2l.load_array((train_features, train_labels.reshape(-1,1)),
                                batch_size)
    test_iter = d2l.load_array((test_features, test_labels.reshape(-1,1)),
                               batch_size, is_train=False)
    trainer = torch.optim.SGD(net.parameters(), lr=0.001)
    animator = d2l.Animator(xlabel='epoch', ylabel='loss', yscale='log',
                            xlim=[1, num_epochs], ylim=[1e-3, 1e2],
                            legend=['train', 'test'])
    for epoch in range(num_epochs):
        train_epoch_ch3(net, train_iter, loss, trainer)
        if epoch == 0 or (epoch + 1) % 20 == 0:
            animator.add(epoch + 1, (evaluate_loss(net, train_iter, loss),
                                     evaluate_loss(net, test_iter, loss)))
    print('weight:', net[0].weight.data.numpy())
```

```python
def train_epoch_ch3(net, train_iter, loss, updater):
    """The training loop defined in Chapter 3."""
    # Set the model to training mode
    if isinstance(net, torch.nn.Module):
        net.train()
    # Sum of training loss, sum of training accuracy, no. of examples
    metric = d2l.Accumulator(3)
    for X, y in train_iter:
        # Compute gradients and update parameters
        y_hat = net(X)
        l = loss(y_hat, y)
        if isinstance(updater, torch.optim.Optimizer):
            # Using PyTorch in-built optimizer & loss criterion
            updater.zero_grad()
            l.sum().backward()
            updater.step()
        else:
            # Using custom built optimizer & loss criterion
            l.sum().backward()
            updater(X.shape[0])
        metric.add(float(l.sum()), d2l.accuracy(y_hat, y), y.numel())
    # Return training loss and training accuracy
    return metric[0] / metric[2], metric[1] / metric[2]
```

## 4.4.4.3 Third-Order Polynomial Function Fitting (Normal)

We will begin by first using a third-order polynomial function, which is the same order as that of the data generation function.
The results show that this model's training and test losses can be both effectively reduced.
The learned model parameters are also close
to the true values $w = [5, 1.2, -3.4, 5.6]$.

```python
# Pick the first four dimensions, i.e., 1, x, x^2/2!, x^3/3! from the
# polynomial features
train(poly_features[:n_train, :4], poly_features[n_train:, :4],
      labels[:n_train], labels[n_train:])
```

## 4.4.4.4. Linear Function Fitting (Underfitting)

Let us take another look at linear function fitting. After the decline in early epochs, it becomes difficult to further decrease this model’s training loss. After the last epoch iteration has been completed, the training loss is still high. When used to fit nonlinear patterns (like the third-order polynomial function here) linear models are liable to underfit.

```python
# Pick the first two dimensions, i.e., 1, x, from the polynomial features
train(poly_features[:n_train, :2], poly_features[n_train:, :2],
      labels[:n_train], labels[n_train:])
```

## 4.4.4.5. Higher-Order Polynomial Function Fitting (Overfitting)

Now let us try to train the model using a polynomial of too high degree. Here, there are insufficient data to learn that the higher-degree coefficients should have values close to zero. As a result, our overly-complex model is so susceptible that it is being influenced by noise in the training data. Though the training loss can be effectively reduced, the test loss is still much higher. It shows that the complex model overfits the data.

```python
# Pick all the dimensions from the polynomial features
train(poly_features[:n_train, :], poly_features[n_train:, :],
      labels[:n_train], labels[n_train:], num_epochs=1500)
```
